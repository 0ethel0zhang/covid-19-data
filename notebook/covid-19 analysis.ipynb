# Inflection Point Analysis
i.e. whether the growth on growth rate has slowed down / reached 0 (which means the growth is constant and thus at the inflection point)

```python
#growth rate
diff_days = pd.DataFrame(counties.groupby("date").cases.sum().diff()).reset_index()
```

```python
diff_days.info()
```

```python
diff_days['pct_change'] = diff_days.cases.pct_change()
```

```python
# plot the growth rate on growth
diff_days.set_index("date")['pct_change'].plot(kind = "bar", figsize = (15,5))
```

**NOTE:** When there is no data plotted for a day that is not the start date, it means that the number of new cases added the day before is the same as before. It does not mean that there are no new cases the day before.

# Cases by states
Only using 20 states that have the most cases to date

```python
worst_20 = counties.groupby("state").cases.sum().sort_values().tail(20).index.tolist()
```

```python
daily_state = pd.DataFrame(counties[counties.state.isin(worst_20)].groupby(["date","state"]).cases.sum().reset_index())
```

## Cases per state over time

```python
plt.figure(figsize = (15, 10))
plt.xticks(rotation=90)
sns.lineplot(data = daily_state, x="date", y="cases", hue = "state", palette = "muted")
```

```python
# calculate growth rate
temp = daily_state.copy()
temp['log_cases'] = temp.cases.apply(np.log)
```

```python
rgb_lst = [[240,163,255],[0,117,220],[153,63,0],[76,0,92],[25,25,25],[0,92,49],[43,206,72],[255,204,153],[128,128,128],[148,255,181],[143,124,0],[157,204,0],[194,0,136],[0,51,128],
           [255,164,5],[255,168,187],[66,102,0],[255,0,16],[94,241,242],[0,153,143],[224,255,102],[116,10,255],[153,0,0],[255,255,128],[255,255,0],[255,80,5]]
# distinguishable 26 colors palette [link](https://graphicdesign.stackexchange.com/questions/3682/where-can-i-find-a-large-palette-set-of-contrasting-colors-for-coloring-many-d)
```

```python
def clamp(x):
  # convert RGB to hex [link](https://stackoverflow.com/questions/3380726/converting-a-rgb-color-tuple-to-a-six-digit-code-in-python)
  return max(0, min(x, 255))
```

```python
hex_lst = list(map(lambda x: "#{0:02x}{1:02x}{2:02x}"\
                           .format(clamp(x[0]), clamp(x[1]), clamp(x[2])), rgb_lst))
```

```python
plt.figure(figsize = (15, 10))
plt.xticks(rotation=90)
sns.lineplot(data = temp, x="date", y="log_cases", hue = "state",
             palette = sns.color_palette(hex_lst, n_colors = 20))
```

Though started at different times, all of these 20 states are experiencing exponential growth in number of cases.

## Cases growth per state over time

```python
# calculate growth rate
temp = daily_state.copy()
temp = temp.merge(temp.groupby(["state", "date"]).cases.sum().pct_change().reset_index()\
           .rename(columns = {"cases":"growth"}),
           on = ["state", "date"])
```

```python
plt.figure(figsize = (15, 10))
plt.xticks(rotation=90)
sns.lineplot(data = temp, x="date", y="growth", hue = "state", 
             palette = sns.color_palette(hex_lst, n_colors=20))
```

As can be seen in the graph, though some places may not have as many cases are other states like NY and California, the growth rate on the cases suggests that there are some outbreaks in early March. <br>
Such states are Massachusetts, Tennessee, Ohio, Colorado, North Carolina. <br>
Two states that are experience especially high growth in number of cases are Louisiana, and Michigan.
